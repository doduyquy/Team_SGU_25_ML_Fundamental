# Trực quan hóa dữ liệu về phân loại chất lương rượu đỏ.

## 1. Chuẩn bị dữu liệu ho trực quan hóa dữ liệu.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Tải dữ liệu
df = pd.read_csv('winequality-red.csv')

# Nhóm theo quality và tính trung bình alcohol

wine_grouped = df.groupby('quality')['alcohol'].mean().reset_index()
wine_sorted = wine_grouped.sort_values('alcohol', ascending=False)

# Kiểm tra dữ liệu
print(wine_sorted.head())
```

# 2 Trực quan hóa


### 2.1 Trực quan hóa với hư viện MATPLOTLIB


```python
# Case 1: Basic
plt.figure(figsize=(12, 6))
x = wine_sorted['quality']
y = wine_sorted['alcohol']
plt.bar(x, y, color='skyblue')
plt.show()
```

**Ý nghĩa**

+ Trục x: Mức chất lượng (quality, từ 3 đến 8).
+ Trục y: Độ cồn trung bình (% thể tích).
+ Biểu đồ này cho thấy mức độ cồn trung bình của rượu ở từng mức chất lượng. Các mức chất lượng cao hơn (7, 8) thường có độ cồn cao hơn, như đã quan sát trong phân tích tương quan trước đó.

```python
# Case 2: Advanced 1
plt.figure(figsize=(12, 6))
plt.bar(x, y, color='lightcoral')
plt.title('Trung bình độ cồn theo chất lượng rượu đỏ', fontsize=15)
plt.xlabel('Chất lượng (Quality)', fontsize=12)
plt.xticks(fontsize=10)
plt.ylabel('Độ cồn trung bình (% thể tích)', fontsize=12)
plt.yticks(fontsize=10)
plt.grid(True, alpha=0.3)
plt.show()
```

**Ý nghĩa:**

+ Tương tự Case 1, nhưng biểu đồ được cải thiện với:

+ Tiêu đề rõ ràng: "Trung bình độ cồn theo chất lượng rượu đỏ".
Nhãn trục x (Chất lượng) và y (Độ cồn trung bình) dễ hiểu.
Thêm lưới (grid) để dễ quan sát giá trị.


+ Biểu đồ này giúp nhà sản xuất rượu thấy rõ rằng các mức chất lượng cao (7, 8) thường có độ cồn trung bình cao hơn, gợi ý rằng độ cồn là yếu tố quan trọng.

```python
# Case 3: Advanced 2
# Tính trung bình volatile acidity theo quality
wine_grouped_va = df.groupby('quality')['volatile acidity'].mean().reset_index()
wine_sorted_va = wine_grouped_va.sort_values('volatile acidity', ascending=False)

# Tạo subplot
fig, ax = plt.subplots(figsize=(20, 8))

# Biểu đồ 1: Độ cồn trung bình
plt.subplot(1, 2, 1)
x = wine_sorted['quality']
y = wine_sorted['alcohol']
plt.bar(x, y, color='lightgreen')
plt.title('Trung bình độ cồn theo chất lượng rượu đỏ', fontsize=16)
plt.xlabel('Chất lượng (Quality)', fontsize=14)
plt.ylabel('Độ cồn trung bình (% thể tích)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, alpha=0.3)

# Biểu đồ 2: Độ axit dễ bay hơi trung bình
plt.subplot(1, 2, 2)
x_va = wine_sorted_va['quality']
y_va = wine_sorted_va['volatile acidity']
plt.bar(x_va, y_va, color='salmon')
plt.title('Trung bình độ axit dễ bay hơi theo chất lượng rượu đỏ', fontsize=16)
plt.xlabel('Chất lượng (Quality)', fontsize=14)
plt.ylabel('Độ axit dễ bay hơi trung bình (g/L)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()  # Đảm bảo các biểu đồ không chồng lấn
plt.show()
```

**Ý nghĩa:**

+ Biểu đồ 1 (trái): Hiển thị độ cồn trung bình theo chất lượng, tương tự Case 2 nhưng trong bố cục subplot.
+ Biểu đồ 2 (phải): Hiển thị độ axit dễ bay hơi trung bình (volatile acidity) theo chất lượng. Vì volatile acidity có tương quan nghịch với quality, bạn sẽ thấy các mức chất lượng thấp (3, 4) có độ axit cao hơn, trong khi chất lượng cao (7, 8) có độ axit thấp hơn.
+ So sánh song song: Việc đặt hai biểu đồ cạnh nhau giúp dễ dàng nhận ra mối quan hệ đối lập giữa alcohol (tăng theo chất lượng) và volatile acidity (giảm theo chất lượng).

### 2.2 Với thư viện SEEBORN

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Tải dữ liệu
df = pd.read_csv('winequality-red.csv')

# Nhóm theo quality và tính trung bình alcohol và volatile acidity
wine_grouped = df.groupby('quality')[['alcohol', 'volatile acidity']].mean().reset_index()
wine_sorted = wine_grouped.sort_values('alcohol', ascending=False)

# Kiểm tra dữ liệu
print(wine_sorted.head())
```

```python
# Case 1: Basic
plt.figure(figsize=(12, 6))
data = wine_sorted
sns.barplot(data=data, x='quality', y='alcohol', color='skyblue')
plt.show()
```

**Ý nghĩa:**

+ Trục x: Mức chất lượng (quality, từ 3 đến 8).
+ Trục y: Độ cồn trung bình (% thể tích).
+ Biểu đồ này cho thấy độ cồn trung bình tăng dần theo chất lượng, với các mức chất lượng cao (7, 8) có độ cồn cao hơn.

```python
# Case 2: Advanced 1
plt.figure(figsize=(12, 6))
data = wine_sorted
ax = sns.barplot(data=data, x='quality', y='alcohol', color='lightcoral')
ax.set_xlabel('Chất lượng (Quality)', fontsize=15)
ax.set_ylabel('Độ cồn trung bình (% thể tích)', fontsize=15)
ax.set_title('Trung bình độ cồn theo chất lượng rượu đỏ', fontsize=20)
plt.show()
```

**Ý nghĩa:**

+ Tương tự Case 1, nhưng được cải tiến với:

+ Tiêu đề rõ ràng: "Trung bình độ cồn theo chất lượng rượu đỏ".
Nhãn trục x và y được đặt tên cụ thể, kích thước chữ lớn hơn.


+ Biểu đồ này giúp nhà sản xuất rượu thấy rõ rằng chất lượng cao thường đi kèm với độ cồn cao hơn.

```python
# Case 3: View multiple perspectives at once
fig, ax = plt.subplots(1, 2, figsize=(20, 8))
data = wine_sorted
sns.set(font_scale=1.5)  # Giảm font scale để phù hợp với kích thước biểu đồ

# Biểu đồ 1: Độ cồn trung bình
ax1 = sns.barplot(data=data, x='quality', y='alcohol', ax=ax[0], color='lightgreen')
ax1.set_xlabel('Chất lượng (Quality)', fontsize=14)
ax1.set_ylabel('Độ cồn trung bình (% thể tích)', fontsize=14)
ax1.set_title('Trung bình độ cồn theo chất lượng rượu đỏ', fontsize=16)

# Biểu đồ 2: Độ axit dễ bay hơi trung bình
ax2 = sns.barplot(data=data, x='quality', y='volatile acidity', ax=ax[1], color='salmon')
ax2.set_xlabel('Chất lượng (Quality)', fontsize=14)
ax2.set_ylabel('Độ axit dễ bay hơi trung bình (g/L)', fontsize=14)
ax2.set_title('Trung bình độ axit dễ bay hơi theo chất lượng rượu đỏ', fontsize=16)

plt.tight_layout()  # Đảm bảo các biểu đồ không chồng lấn
plt.show()
```
